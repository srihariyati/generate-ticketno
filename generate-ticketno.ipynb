```python
import pandas as pd
from PIL import Image, ImageFont, ImageDraw
```

```python
#VIP Ticket 0001 - 0500
FONT_FILE = ImageFont.truetype(r'montserrat.bold.ttf', 20)
FONT_COLOR_RIGHT = "#BE9D00"
FONT_COLOR_LEFT =  "#FFFFFF"
# https://www.image-map.ne
WIDTH_RIGHT, HEIGHT_RIGHT, WIDTH_LEFT, HEIGHT_LEFT= 1500,481,393,438 

def make_ticket(number):
    image_source = Image.open(r'infest_ticket_vip.png')
    draw = ImageDraw.Draw(image_source)
    
    #right side
    no_ticket_width_right, no_ticket_height_right = draw.textsize(number, font=FONT_FILE)
    draw.text((WIDTH_RIGHT-no_ticket_width_right/2, HEIGHT_RIGHT-no_ticket_height_right/2),number, fill=FONT_COLOR_RIGHT, font=FONT_FILE)
    
    #left side
    no_ticket_width_left, no_ticket_height_left = draw.textsize(number, font=FONT_FILE)
    draw.text((WIDTH_LEFT-no_ticket_width_left/2, HEIGHT_LEFT-no_ticket_height_left/2),number, fill=FONT_COLOR_LEFT, font=FONT_FILE)
    
    image_source.save("VIP/"+number+".png")
    
    print("Printing VIP : #"+number)
    
for x in range(500):
    x = ("%04d" %int(x+1))
    make_ticket(x)
```
